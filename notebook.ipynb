```python
# Start coding here! Use as many cells as you like

import pandas as pd

# Load the Netflix data from a CSV file
netflix_data = pd.read_csv('netflix_data.csv')

# Print the dataframe
print(netflix_data)
print(netflix_data.info())

# Filter data for movies released in the 1990s
movies_1990s = netflix_data[(netflix_data['release_year'] >= 1990) & 
                            (netflix_data['release_year'] < 2000) & 
                            (netflix_data['type'] == 'Movie')]

# Find the most frequent movie duration in the 1990s
duration = movies_1990s['duration'].mode().iloc[0]

# Count the number of short action movies released in the 1990s
short_movie_count = len(movies_1990s[(movies_1990s['duration'] < 90) & 
                                     (movies_1990s['genre'].str.contains('Action'))])

print(f"Most frequent movie duration in the 1990s: {duration} minutes")
print(f"Number of short action movies released in the 1990s: {short_movie_count}")
```
